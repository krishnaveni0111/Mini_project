```python
from keras.models import load_model
import numpy as np
from keras.preprocessing import image


model=load_model('200epochsModel.h5')
test_image = image.load_img('Test images/asd/KKI_0050774_lfcd.nii.gz.png', target_size = (64, 64))
# Loading the image and converting the pixels into array whcih will be used as input to predict.
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = model.predict(test_image)
#training_set.class_indices
print(result[0][0])
if result[0][0]<0.5:
    prediction = 'asd'
else:
    prediction = 'tdc'
print(prediction)
```
